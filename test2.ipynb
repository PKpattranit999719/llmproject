# แก้ปัญหา สร้างฟังก์ชันการกรองข้อมูลจาก DataFrame ตามคอลัมน์ที่กำหนด: โดยใช้ filter เพื่อเลือกหรือกรองข้อมูลจาก DataFrame ที่เรามี

```python
def filter_data(df: pd.DataFrame, thai_name: str = None, region: str = None, subdistrict: str = None,
                district: str = None, province: str = None, category: str = None, subtype: str = None) -> pd.DataFrame:
    """
    ฟังก์ชันนี้กรองข้อมูลจาก DataFrame ตามค่าที่ระบุในคอลัมน์ต่างๆ
    """
    # ตรวจสอบ DataFrame ก่อนกรอง
    if df.empty:
        print("DataFrame is empty!")
        return pd.DataFrame()

    print("Initial DataFrame shape:", df.shape)

    # กรองข้อมูลตามคอลัมน์
    if thai_name:
        df = df[df['ATT_NAME_TH'].str.contains(thai_name, na=False, case=False)]
    if region:
        df = df[df['REGION_NAME_TH'].str.contains(region, na=False, case=False)]
    if subdistrict:
        df = df[df['SUBDISTRICT_NAME_TH'].str.contains(subdistrict, na=False, case=False)]
    if district:
        df = df[df['DISTRICT_NAME_TH'].str.contains(district, na=False, case=False)]
    if province:
        df = df[df['PROVINCE_NAME_TH'].str.contains(province, na=False, case=False)]
    if category:
        df = df[df['ATTR_CATAGORY_TH'].str.contains(category, na=False, case=False)]
    if subtype:
        df = df[df['ATTR_SUB_TYPE_TH'].str.contains(subtype, na=False, case=False)]

    # ตรวจสอบผลลัพธ์หลังการกรอง
    if df.empty:
        print("No matching records found.")
    
    print("Filtered DataFrame shape:", df.shape)
    return df
```

```python
response = agent.run("ช่วยหาอ่างเก็บน้ำลำพอก จากข้อมูลในไฟล์splitData.csvให้หน่อย")
print(response)
```

```python
##สร้าง Custom Tool สำหรับการกรองข้อมูล:
from langchain.agents import Tool

# สร้าง Custom Tool
filter_data_tool = Tool(
    name="Filter Data",
    func=filter_data,  # ฟังก์ชันที่สร้างไว้
    description="กรองข้อมูลจาก DataFrame ตามคอลัมน์ต่างๆ เช่น ATT_NAME_TH, REGION_NAME_TH ฯลฯ"
)
```

```python
## เพิ่ม Custom Tool เข้าไปใน Agent
from langchain.agents import initialize_agent
from langchain_openai import ChatOpenAI

llm = ChatOpenAI(
    model='gpt-4o-mini',
    base_url= os.getenv("url"),  
    api_key= os.getenv("api_key"),  
    max_tokens=1000 
)

# สร้าง Agent โดยใช้ LLM ที่คุณต้องการ
tools = [filter_data_tool]  # เพิ่ม tool ที่สร้างไว้ในรายการเครื่องมือที่ Agent สามารถใช้ได้

# สร้าง Agent โดยใช้ฟังก์ชันที่เป็น tool
agent = initialize_agent(
    tools=tools, 
    llm=llm,  # LLM ของคุณ
    agent_type="zero-shot-react-description",
    verbose=True
)

# ตั้งคำถามให้ Agent
# แสดงผลลัพธ์
response = agent.run("แหล่งท่องเที่ยวทางธรรมชาติ แบบหมู่เกาะ จากข้อมูลในไฟล์splitData.csvให้หน่อย")
print(response)
```

# ทำการทดลองให้การ QUERY แม่นยิ่งขึ้น

```python
from langchain_experimental.agents import create_pandas_dataframe_agent
from langchain_openai import ChatOpenAI
import pandas as pd

# สร้างโมเดลที่ใช้เชื่อมต่อกับเซิร์ฟเวอร์กลาง
llm = ChatOpenAI(
    model='gpt-4o-mini',
    base_url= os.getenv("url"),  
    api_key= os.getenv("api_key"),  
    max_tokens=1000 
)

# โหลดข้อมูล
df = pd.read_csv("splitData.csv")

def extract_keywords_from_query(query):
    # ส่งคำถามไปยังเซิร์ฟเวอร์กลางเพื่อให้ ChatGPT ช่วยแยกคำสำคัญ
    response = llm.invoke(f"แยกคำสำคัญจากคำถามนี้: '{query}' และแสดงคำสำคัญที่เกี่ยวข้องที่สามารถใช้กรองข้อมูลได้")
    
    # ตรวจสอบและดึงข้อความจาก AIMessage
    return response.content.strip()  # ใช้ .content แทน .text


# ตัวอย่างคำถามจากผู้ใช้
user_query = "ช่วยหาแหล่งท่องเที่ยวทางธรรมชาติ เสาหินบะซอลต์ ภาคตะวันออกเฉียงเหนือ"

# ใช้ ChatGPT เพื่อแยกคำสำคัญจากคำถาม
keywords = extract_keywords_from_query(user_query)
print("คำสำคัญที่ได้จากการแยกคำ:", keywords)

# ฟังก์ชันกรองข้อมูลจาก DataFrame
def filter_data_with_keywords(df, keywords):
    # แปลงคำสำคัญที่ได้จาก ChatGPT ไปใช้ในการกรองข้อมูล
    filters = keywords.split(",")
    for keyword in filters:
        df = df[df.apply(lambda row: row.astype(str).str.contains(keyword, case=False, na=False).any(), axis=1)]
    return df

# กรองข้อมูลจาก DataFrame ด้วยคำสำคัญที่ได้จาก ChatGPT
filtered_df = filter_data_with_keywords(df, keywords)

# แสดงผลลัพธ์หลังการกรอง
print(filtered_df)
```

ผลทีได้มี 2 ส่วนด้วยกัน คำสำคัญที่ได้จากการแยกคำ อันนี้ถือว่าแยกได้ดี
แต่ส่วนที่2 ในส่วนของการ ผลลัพธ์จากการกรองข้อมูล 
Empty DataFrame: หมายความว่าไม่มีข้อมูลใน DataFrame ที่ตรงกับคำสำคัญที่แยกออกมา ก็คือหาไม่เจอนั้นแหละ

```python
def extract_keywords_from_query(query: str) -> str:
    """
    ใช้โมเดล LLM เพื่อแยกคำสำคัญจากคำถามของผู้ใช้
    """
    # ส่งคำถามไปยังโมเดล LLM
    response = llm.invoke(f"แยกคำสำคัญจากคำถามนี้: '{query}' และแสดงคำสำคัญที่เกี่ยวข้องที่สามารถใช้กรองข้อมูลได้")
    
    # ตรวจสอบผลลัพธ์และคืนค่าคำสำคัญ
    return response.content.strip()
```

```python
def filter_data(df: pd.DataFrame, thai_name: str = None, region: str = None, subdistrict: str = None,
                district: str = None, province: str = None, category: str = None, subtype: str = None) -> pd.DataFrame:
    """
    ฟังก์ชันกรองข้อมูลจาก DataFrame ตามค่าที่ระบุในคอลัมน์ต่างๆ
    """
    if df.empty:
        print("DataFrame is empty!")
        return pd.DataFrame()

    # กรองข้อมูลตามคอลัมน์
    if thai_name:
        df = df[df['ATT_NAME_TH'].str.contains(thai_name, na=False, case=False)]
    if region:
        df = df[df['REGION_NAME_TH'].str.contains(region, na=False, case=False)]
    if subdistrict:
        df = df[df['SUBDISTRICT_NAME_TH'].str.contains(subdistrict, na=False, case=False)]
    if district:
        df = df[df['DISTRICT_NAME_TH'].str.contains(district, na=False, case=False)]
    if province:
        df = df[df['PROVINCE_NAME_TH'].str.contains(province, na=False, case=False)]
    if category:
        df = df[df['ATTR_CATAGORY_TH'].str.contains(category, na=False, case=False)]
    if subtype:
        df = df[df['ATTR_SUB_TYPE_TH'].str.contains(subtype, na=False, case=False)]

    return df
```

# ลองแก้ให้ผลลัพย์สอดคล้องกับสิ่งที่ userQuery  

```python
# 1
from langchain_experimental.agents import create_pandas_dataframe_agent
from langchain_openai import ChatOpenAI
import pandas as pd

# สร้างโมเดลที่ใช้เชื่อมต่อกับเซิร์ฟเวอร์กลาง
llm = ChatOpenAI(
    model='gpt-4o-mini',
    base_url= os.getenv("url"),  
    api_key= os.getenv("api_key"),  
    max_tokens=1000 
)

# โหลดข้อมูล
df = pd.read_csv("splitData.csv")  # โหลดไฟล์ข้อมูล
```

```python
# 2
def extract_keywords_from_query(query: str) -> dict:
    """
    ใช้โมเดล LLM เพื่อแยกคำสำคัญจากคำถามของผู้ใช้และระบุคอลัมน์ที่เกี่ยวข้องสำหรับการกรอง
    """
    response = llm.invoke(
        f"""
        วิเคราะห์คำถาม: '{query}'
        ระบุคำสำคัญที่เกี่ยวข้องสำหรับการกรองข้อมูลใน DataFrame โดยแยกคำสำคัญออกเป็น
        คอลัมน์ที่เหมาะสม: ATT_NAME_TH, REGION_NAME_TH, PROVINCE_NAME_TH, ATTR_CATAGORY_TH, เป็นต้น
        ส่งออกในรูปแบบ JSON เช่น:
        {{
            "ATT_NAME_TH": ["คำสำคัญ1", "คำสำคัญ2"],
            "PROVINCE_NAME_TH": ["คำสำคัญ3"]
        }}
        """
    )
    return response.content.strip()
```

```python
# 3
def filter_data(df: pd.DataFrame, thai_name: str = None, region: str = None, subdistrict: str = None,
                district: str = None, province: str = None, category: str = None, subtype: str = None) -> pd.DataFrame:
    """
    ฟังก์ชันกรองข้อมูลจาก DataFrame ตามค่าที่ระบุในคอลัมน์ต่างๆ
    """
    if df.empty:
        print("DataFrame is empty!")
        return pd.DataFrame()

    print("Initial DataFrame shape:", df.shape)
    
    # กรองข้อมูลตามคอลัมน์
    if thai_name:
        df = df[df['ATT_NAME_TH'].str.contains(thai_name, na=False, case=False)]
    if region:
        df = df[df['REGION_NAME_TH'].str.contains(region, na=False, case=False)]
    if subdistrict:
        df = df[df['SUBDISTRICT_NAME_TH'].str.contains(subdistrict, na=False, case=False)]
    if district:
        df = df[df['DISTRICT_NAME_TH'].str.contains(district, na=False, case=False)]
    if province:
        df = df[df['PROVINCE_NAME_TH'].str.contains(province, na=False, case=False)]
    if category:
        df = df[df['ATTR_CATAGORY_TH'].str.contains(category, na=False, case=False)]
    if subtype:
        df = df[df['ATTR_SUB_TYPE_TH'].str.contains(subtype, na=False, case=False)]

    return df
```

```python
# 4
from langchain.agents import Tool

# สร้าง Custom Tool
filter_data_tool = Tool(
    name="Filter Data",
    func=filter_data,
    description="กรองข้อมูลจาก DataFrame ตามคอลัมน์ต่างๆ เช่น ATT_NAME_TH, REGION_NAME_TH ฯลฯ"
)
```

```python
# 5
from langchain_experimental.agents.agent_toolkits import create_pandas_dataframe_agent

agent = create_pandas_dataframe_agent(
    llm=llm,
    df=df,
    verbose=True,
    allow_dangerous_code=True,  # ใช้ตัวเลือกนี้อย่างระมัดระวัง
    agent_type="tool-calling",
    tools=[filter_data_tool]
)
```

```python
import pandas as pd
from io import StringIO

# ตัวอย่างคำถามของผู้ใช้
user_query = "แหล่งท่องเที่ยวสำหรับกิจกรรมพิเศษนันทนาการและความสนใจพิเศษในจังหวัดประจวบคีรีขันธ์ให้หน่อย"

# ใช้ฟังก์ชันแยกคำสำคัญ
keywords = extract_keywords_from_query(user_query)
print("คำสำคัญที่ได้จากคำถาม:", keywords)

# ตรวจสอบว่า keywords เป็นพจนานุกรมหรือไม่
if isinstance(keywords, dict):
    # ใช้คำสำคัญสร้างเงื่อนไขสำหรับการกรอง
    filter_conditions = []
    for column, values in keywords.items():
        condition = f"{column}.str.contains('|'.join({values}), na=False, case=False)"
        filter_conditions.append(condition)

    filter_query = " & ".join(filter_conditions)

    # ใช้ Agent รันคำสั่งกรองข้อมูล
    response = agent.run(f"""
    กรองข้อมูลใน DataFrame โดยใช้เงื่อนไข:
    {filter_query}
    แสดงเฉพาะชื่อสถานที่ (ATT_NAME_TH), จังหวัด (PROVINCE_NAME_TH), และประเภทสถานที่ (ATTR_CATAGORY_TH)
    ในรูปแบบตาราง
    """)
    print("ผลลัพธ์ที่ได้จาก Agent:", response)

    # ตรวจสอบว่า response เป็นข้อความหรือไม่ ถ้าเป็นข้อความ CSV ก็แปลงเป็น DataFrame
    if isinstance(response, str):
        # สมมติว่า response เป็นข้อความ CSV
        df = pd.read_csv(StringIO(response))
        print("DataFrame หลังจากแปลงจากข้อความ CSV:")
        print(df)

        # กรองข้อมูลจาก DataFrame
        filtered_df = df.query(filter_query)
        print("ข้อมูลที่กรองแล้ว:")
        print(filtered_df)
    else:
        print("ข้อมูลที่ได้ไม่เป็นข้อความที่สามารถแปลงได้")
else:
    print("คำสำคัญที่ได้ไม่เป็นพจนานุกรม (dict) กรุณาตรวจสอบฟังก์ชัน extract_keywords_from_query")
```
